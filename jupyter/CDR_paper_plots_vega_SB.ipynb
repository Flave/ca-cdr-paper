```python
import pandas as pd
import altair as alt
from altair import datum
```

```python
data = data = pd.read_csv(
  "https://raw.githubusercontent.com/Flave/ca-cdr-paper/master/paper-data.csv"
,parse_dates=['year'], 
)
```

```python
data
```

```python
#POP lines

CDR_mean = alt.Chart(data
).transform_filter(
  'datum.region!="World"'
).transform_filter(
  'datum.region!="MEA"'
).transform_calculate(
  per_capita='datum.cum_co2/datum.cum_population'
).encode(
  x='year',
  y=alt.Y('mean(per_capita):Q', title="Cumulative emissions per capita (GtCO2e/yr)"),
  color=alt.Color('region:N', sort='descending')
).mark_line().properties(title='Equity-based CRD distributions')


# plot ranges of mean
CDR = alt.Chart(data
).transform_filter(
  'datum.region!="World"'
).transform_calculate(
  per_capita='datum.cum_co2/datum.cum_population'
).encode(
  x='year',
  y=alt.Y('per_capita:Q', title="Cumulative emissions per capita (GtCO2e/yr)"),
  color=alt.Color('region:N', sort='descending')
).mark_area().properties(title='Equity-based CDR distributions')

ssp1 = CDR.transform_filter(alt.FieldEqualPredicate(field='scenario', equal='SSP1-19-SPA0-V17'))


#GDP lines

GDP_mean = alt.Chart(data
).transform_filter(
  'datum.region!="World"'
).transform_filter(
  'datum.region!="MEA"'
).encode(
  x='year',
  y=alt.Y('mean(gdp_cap):Q', title="GDP per capita (thousand US$ yrâ€“1)"),
  color=alt.Color('region:N', sort='descending')
).mark_line().properties(title='Equity-based CDR distributions')


CDR_mean | GDP_mean
```

```python
mouse_selection = alt.selection(type='single', fields=['region', 'year'], on='mouseover', clear='mouseout', nearest=True)

#POP lines

base = alt.Chart(data
).transform_filter(
  'datum.region!="World" && datum.region!="MEA"'
).transform_calculate(
  per_capita='datum.cum_co2/datum.cum_population'
).encode(
  x='year',
  y=alt.Y('mean(per_capita):Q', title="Cumulative emissions per capita (GtCO2e/yr)"),
  color=alt.Color(
    'region:N', 
    scale = alt.Scale(
      domain = ['CHN', 'IND', 'RUS', 'USA', 'AFR', 'LAM', 'EUR'], 
      range = [ '#b2df8a','#ffff99', '#1f78b4', '#ff7f00','#fb9a99', '#ffd92f', '6a3d9a']),
    sort='descending'
  )
)

#b3b3b3
CDR_mean = base.mark_line().properties(title='Equity-based CDR distributions')

CDR_dots = base.mark_circle().properties(title='Equity-based CDR distributions').encode(
    tooltip=['region','mean(per_capita):Q', 'year'],
    size=alt.condition(mouse_selection, alt.value(70), alt.value(20))
).add_selection(mouse_selection)

CDR_all = alt.Chart(data
).transform_filter(
  'datum.region!="World"'
).transform_filter(
  'datum.region!="MEA"'
).transform_calculate(
  per_capita='datum.cum_co2/datum.cum_population'
).encode(
  x='year',
  y=alt.Y('per_capita:Q', title="Cumulative emissions per capita (GtCO2e/yr)"),
  detail = ['region:N', 'scenario:N']
).mark_line(color="rgba(0, 0, 0, 0.05)").properties(title='Equity-based CDR distributions')

# Why can't I do CDR_all lines in gray?
# Why does mouse selection not work?
#_________________________________________________________?



CDR_all + CDR_mean + CDR_dots

```

```python
region_dropdown = alt.binding_select(options = [None, 'CHN', 'IND', 'RUS', 'USA', 'AFR', 'LAM', 'EUR'])
region_selection = alt.selection_single(fields=['region'], bind=region_dropdown, init={'region': 'USA'})

mouse_selection = alt.selection(type='single', on='mouseover', clear='mouseout', nearest=True)

#POP lines

base = alt.Chart(data
).transform_filter(
  'datum.region!="World"'
).transform_filter(
  'datum.region!="MEA"'
).transform_calculate(
  per_capita='datum.cum_co2/datum.cum_population'
).encode(
  x='year',
  y=alt.Y('mean(per_capita):Q', title="Cumulative emissions per capita (GtCO2e/yr)"),
  color=alt.condition(region_selection, alt.Color('region:N', 
                                                  scale = alt.Scale(domain = ['CHN', 'IND', 'RUS', 'USA', 'AFR', 'LAM', 'EUR'],
                                                range = [ '#b2df8a','#ffff99', '#1f78b4',
                                                          '#ff7f00','#fb9a99', '#ffd92f', '6a3d9a'])),
                      alt.value('rgba(0,0,0,0.05)')), detail='region:N')


CDR_all = alt.Chart(data
).transform_filter(
  'datum.region!="World"'
).transform_filter(
  'datum.region!="MEA"'
).transform_calculate(
  per_capita='datum.cum_co2/datum.cum_population'
).encode(
  x='year',
  y=alt.Y('per_capita:Q', title="Cumulative emissions per capita (GtCO2e/yr)"),
  color=alt.value('#999'), 
  detail = 'scenario'
).mark_line().properties(title='Equity-based CDR distributions')


# Why can't I do CDR_all lines in gray?
# Why does mouse selection not work?
#_________________________________________________________?

CDR_mean = base.mark_line().properties(title='Equity-based CDR distributions').add_selection(region_selection)

CDR_dots = base.mark_circle().properties(title='Equity-based CDR distributions').transform_filter(region_selection).encode(
    tooltip=['region','mean(per_capita):Q', 'year'],
    size=alt.condition(mouse_selection, alt.value(70), alt.value(20))
).add_selection(mouse_selection)



CDR_mean + CDR_dots# + CDR_all
```
